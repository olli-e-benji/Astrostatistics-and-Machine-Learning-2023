# Pdf of the log

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
min = 0.1
max = 10
N = 5000

samples = np.random.uniform(min, max, N)

#Raw Histogram
plt.title('Samples from uniform distribution between {} and {}'.format(min, max))
plt.hist(samples, bins=10, density=True, label='Drawn Data')
plt.legend()
plt.show()

y = np.log10(samples)

#Deriving pdf for y
y_axis = np.linspace(np.log10(min), np.log10(max), 1000)
pdf_y = 10**y_axis * np.log(10)/(max-min)

#Log10 Histogram with theoretical pdf overplot
plt.title('Log10 of samples')
plt.hist(y, bins=10, density=True, label='Log10 Data')
plt.plot(y_axis, pdf_y, c='red', label='Theoretical PDF')
plt.legend()
plt.show()

#Verifing robustness of cumulative statistics
sample_mean = np.log10(np.mean(samples))
log10_mean = np.mean(y)
sample_median = np.log10(np.median(samples))
log10_median = np.median(y)

print('Log10 of mean of samples: {}\nMean of Log10 of samples: {}'.format(sample_mean, log10_mean))
print('Log10 of median of samples: {}\nMedian of Log10 of samples: {}'.format(sample_median, log10_median))
```
