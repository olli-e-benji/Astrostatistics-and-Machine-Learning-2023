# Position of a Quasar

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
mu = 1
sigma = 0.2
N = 10

positions = norm.rvs(mu, sigma, N)

centers = []

x_axis = np.linspace(0, 2, 1000)

for i in range(N):
    temporary = norm(positions[i], sigma)
    centers.append(temporary.pdf(x_axis))
    plt.plot(x_axis, temporary.pdf(x_axis))

centers = np.array(centers)

likelihood = np.prod(centers)

print(likelihood)
print(np.mean(centers))
```
