# Horsekicks

<center>

| Number of deaths | Number of groups |
| --- | --- |
| 0 | 109 |
| 1 | 65 |
| 2 | 22 |
| 3 | 3 |
| 4 | 1 |

</center>

```python
import numpy as np
from scipy.stats import poisson as ps
import matplotlib.pyplot as plt
```

```python
deaths = np.array([0, 1, 2, 3, 4])
groups = np.array([109, 65, 22, 3, 1])

#Normalization
all_deaths = np.sum(groups)
density = groups/all_deaths

#Weighted average -> Sensible poissonian distribution
mu = np.average(deaths, weights=density)
distribution = ps(mu).pmf(deaths)

#Plotting
plt.xlabel('Deaths')
plt.ylabel('Group density')
plt.scatter(deaths, density, c='blue', label='Data points')
plt.plot(deaths, distribution, c='orange', label='Theoretical distribution')
plt.legend()
plt.show()
```
